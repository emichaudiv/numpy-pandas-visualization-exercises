```python
import pandas as pd
import numpy as np
from pydataset import data
np.random.seed(123)
```

```python
students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas', 'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

math_grades = np.random.randint(low=60, high=100, size=len(students))

english_grades = np.random.randint(low=60, high=100, size=len(students))

reading_grades = np.random.randint(low=60, high=100, size=len(students))

df_dict = {'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades}

df = pd.DataFrame(df_dict)
```

```python
df['Passing_English'] = df.english >=70
```

```python
df
```

```python
pass_english = df.sort_values(by = 'Passing_English')
pass_english
```

```python
passing_english = df.sort_values(by = 'name').sort_values(by = 'Passing_English')
passing_english
```

```python
grade_english = df.sort_values(by = 'Passing_English').sort_values(by = 'english',ascending=False)
grade_english
```

```python
df['overall_grade']=((df.math+df.english+df.reading)/3).round(2)
df
```
