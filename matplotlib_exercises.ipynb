```python
from matplotlib import pyplot as plt
import pandas as pd
from pydataset import data
import numpy as np
import random
import math
np.random.seed(123)
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
xline = np.array([1,2,3])
yline = xline**2 -xline+2
plt.plot(xline, yline)
plt.show()
```

```python
xsquare = np.array([1,2,3])
ysquare = (xsquare**(1/2))
plt.plot(xsquare, ysquare)
plt.show()
```

```python
xcube = np.array([1,2,3])
ycube = xcube**3
plt.plot(xcube, ycube)
plt.show()
```

```python
x2 = np.array([1,2,3])
y2 = 2**x2
plt.plot(x2, y2)
plt.show()
```

```python
xmore = np.array([1,2,3])
ymore = 1/(xmore+1)
plt.plot(xmore, ymore)
plt.show()
```
